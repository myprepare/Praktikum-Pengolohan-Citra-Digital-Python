## Import Library

```python
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
import numpy as np

import cv2
```

## Load Image

```python
citra1 = imread(fname="mikro gray.jpg")
print(citra1.shape)

plt.imshow(citra1, cmap='gray')
```

## Low Pass Filter

```python
kernel1 = np.array([[-1, -1, -1], 
                   [-1, 9, -1], 
                   [-1, -1, -1]])

kernel2 = np.array([[-1, -1, -1], 
                   [-1, 10, -1], 
                   [-1, -1, -1]])

kernel3 = np.array([[-1, -1, -1], 
                   [-1, 11, -1], 
                   [-1, -1, -1]])

citraOutput1 = cv2.filter2D(citra1, -1, kernel1)
citraOutput2 = cv2.filter2D(citra1, -1, kernel2)
citraOutput3 = cv2.filter2D(citra1, -1, kernel3)

fig, axes = plt.subplots(4, 2, figsize=(20, 20))
ax = axes.ravel()

ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Citra Input")
ax[1].hist(citra1.ravel(), bins = 256)
ax[1].set_title("Histogram Citra Input")

ax[2].imshow(citraOutput1, cmap = 'gray')
ax[2].set_title("Citra Output 1")
ax[3].hist(citraOutput1.ravel(), bins = 256)
ax[3].set_title("Histogram Citra Output 1")

ax[4].imshow(citraOutput2, cmap = 'gray')
ax[4].set_title("Citra Output 2")
ax[5].hist(citraOutput2.ravel(), bins = 256)
ax[5].set_title("Histogram Citra Output 2")

ax[6].imshow(citraOutput3, cmap = 'gray')
ax[6].set_title("Citra Output 3")
ax[7].hist(citraOutput3.ravel(), bins = 256)
ax[7].set_title("Histogram Citra Output 3")

fig.tight_layout()

#plt.savefig("Low Pass.jpg")
```
