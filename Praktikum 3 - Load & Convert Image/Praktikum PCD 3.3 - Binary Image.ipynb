# Import Library

```python
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.pyplot as plt
from skimage import data
```

# Percobaan 1 - Binary Image (Thresholding)
1. Sumber skimage data :https://scikit-image.org/docs/dev/api/skimage.data.html
2. Sumber Image Thresholding : https://scikit-image.org/docs/dev/auto_examples/segmentation/plot_thresholding.html
3. Threshold Mean : https://scikit-image.org/docs/dev/api/skimage.filters.html#skimage.filters.threshold_mean
4. Threshold Minimum : https://scikit-image.org/docs/dev/api/skimage.filters.html#skimage.filters.threshold_minimum

```python
from skimage.filters import threshold_otsu
from skimage.filters import threshold_minimum
from skimage.filters import threshold_mean

image = data.camera()

threshMin = threshold_minimum(image)
print('Threshold Minimum : ',threshMin)
minimumImage = image > threshMin

threshMean = threshold_mean(image)
print('Threshold Mean : ',threshMean)
meanImage = image > threshMean
```

# Percobaan 2 - Plot Hasil Thresholding

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].set_title("Original")
ax[1].hist(image.ravel(), bins=256)
ax[1].set_title('Histogram')

ax[2].imshow(minimumImage, cmap=plt.cm.gray)
ax[2].set_title('Thresholded (min)')
ax[3].hist(image.ravel(), bins=256)
ax[3].axvline(threshMin, color='r')
ax[3].set_title('Histogram (min)')

ax[4].imshow(meanImage, cmap=plt.cm.gray)
ax[4].set_title('Thresholded (mean)')
ax[5].hist(image.ravel(), bins=256)
ax[5].axvline(threshMean, color='r')
ax[5].set_title('Histogram (mean)')

plt.tight_layout()
```
