# Import Library

```python
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
```

# Percobaan 1 - Load From skimage.data
Sumber skimage data = https://scikit-image.org/docs/dev/api/skimage.data.html

```python
original = data.astronaut()
print(original.shape)

plt.imshow(original)
```

# Percobaan 2 - Mengakses Channel RGB (versi 1)

```python
red = original[:,:,0]
green = original[:,:,1]
blue = original[:,:,2]

print('Red Shape = ',red.shape)
print('Green Shape = ',green.shape)
print('Blue Shape = ',blue.shape)

fig, axes = plt.subplots(2, 2, figsize=(10, 10))
ax = axes.ravel()

ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].imshow(original[:,:,0], cmap=plt.cm.gray)
ax[1].set_title("Red Channel")
ax[2].imshow(original[:,:,1], cmap=plt.cm.gray)
ax[2].set_title("Green Channel")
ax[3].imshow(original[:,:,2], cmap=plt.cm.gray)
ax[3].set_title("Blue Channel")
fig.tight_layout()
plt.show()
```

# Percobaan 3 - Mengakses Channel RGB (versi 2)

```python
red = original.copy() #red channel [0]
red[:,:,1]=0
red[:,:,2]=0
print('Red Shape = ',red.shape)

green = original.copy() #green channel [1]
green[:,:,0]=0
green[:,:,2]=0
print('Green Shape = ',green.shape)

blue = original.copy() #green channel [2]
blue[:,:,0]=0
blue[:,:,1]=0
print('Blue Shape = ',blue.shape)

fig, axes = plt.subplots(2, 2, figsize=(10, 10))
ax = axes.ravel()
ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].imshow(red)
ax[1].set_title("Red Channel")
ax[2].imshow(green)
ax[2].set_title("Green Channel")
ax[3].imshow(blue)
ax[3].set_title("Blue Channel")
fig.tight_layout()
plt.show()
```
