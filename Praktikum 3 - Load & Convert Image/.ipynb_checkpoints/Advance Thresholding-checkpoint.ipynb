```python
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.pyplot as plt
from skimage import data
from skimage.filters import threshold_otsu
```

Sumber skimage data :https://scikit-image.org/docs/dev/api/skimage.data.html

Sumber Image Thresholding : https://scikit-image.org/docs/dev/auto_examples/segmentation/plot_thresholding.html

```python
from skimage.filters import threshold_minimum
from skimage.filters import threshold_mean

image = data.camera()

threshMin = threshold_minimum(image)
binaryMin = image > threshMin

threshMean = threshold_mean(image)
binaryMean = image > threshMean
```

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].set_title("Original")
ax[1].hist(image.ravel(), bins=256)
ax[1].set_title('Histogram')

ax[2].imshow(binaryMin, cmap=plt.cm.gray)
ax[2].set_title('Thresholded (min)')
ax[3].hist(image.ravel(), bins=256)
ax[3].axvline(threshMin, color='r')
ax[3].set_title('Histogram (min)')

ax[4].imshow(binaryMean, cmap=plt.cm.gray)
ax[4].set_title('Thresholded (min)')
ax[5].hist(image.ravel(), bins=256)
ax[5].axvline(threshMean, color='r')
ax[5].set_title('Histogram (min)')

plt.tight_layout()
```
