# Import Library

```python
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
import numpy as np
```

# Load Image

```python
citra1 = imread(fname="gedung.tif")
print(citra1.shape)

plt.imshow(citra1, cmap='gray')
```

# Proses Konvolusi

```python
from scipy import ndimage
from skimage.filters.edges import convolve
kernel = np.array([[-1, 0, -1],[0, 4, 0],[-1, 0, -1]])
#output = ndimage.convolve(citra1, kernel, mode='constant', cval=0.0)
output = convolve(citra1, kernel, mode='constant', cval=0.5)

plt.imshow(output, cmap='gray')
```
