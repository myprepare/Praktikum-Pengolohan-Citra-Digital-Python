# Import Library

```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

# Percobaan 1 - Line Plot

```python
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

plt.figure()
plt.plot(t, s)
plt.title('Plot Nilai Fungsi 2*Sin(t1)')
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo (mV)')
plt.grid()
plt.show()
```

# Percobaan 2 - Line Plot dengan 'figsize'

```python
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

plt.figure(figsize=(10,5))
plt.plot(t, s)
plt.title('Plot Nilai Fungsi 2*Sin(t1)')
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo (mV)')
plt.grid()
plt.show()
```

# Percobaan 3 - Bar Chart (Vertical)

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]

plt.figure(figsize=(10,5))
plt.bar(langs,students)
plt.ylabel('Pengguna')
plt.xlabel('Bahasa Pemrograman')
plt.show()
```

# Percobaan 4 - Bar Chart (Horizontal)

```python
y_pos = np.arange(len(langs))

plt.figure(figsize=(10,5))
plt.barh(y_pos, students)
plt.yticks(y_pos, langs)
plt.xlabel('Pengguna')
plt.ylabel('Bahasa Pemrograman')
plt.title('Programming language usage')
plt.show()
```

# Percobaan 5 - Pie Chart

```python
plt.figure(figsize=(10,5))
plt.pie(students, labels = langs,autopct='%1.2f%%')
plt.show()
```

# Percobaan 6 - Pie Chart dengan 'explode'

```python
explode = (0, 0.1, 0, 0, 0)

plt.figure(figsize=(10,5))
plt.pie(students, explode=explode, labels = langs,autopct='%1.2f%%')
plt.show()
```

# Percobaan 7 - 2 Subplot 

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.figure(figsize=(10,5))
plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title('Subplot ke 1')
plt.ylabel('Amplitudo (mV)')
plt.xlabel('time (s)')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(x2, y2)
plt.title('Subplot ke 2')
plt.xlabel('time (ms)')
plt.ylabel('Amplitudo (mV)')
plt.tight_layout()
plt.show()
```

# Percobaan 8 - 4 Subplot

```python
plt.figure(figsize=(10,10))
plt.subplot(2, 2, 1)
plt.pie(students, labels = langs,autopct='%1.2f%%')
plt.title('Subplot ke 1')

plt.subplot(2, 2, 2)
explode = (0, 0.1, 0, 0, 0)
plt.pie(students, explode=explode, labels = langs,autopct='%1.2f%%')
plt.title('Subplot ke 2')

plt.subplot(2, 2, 3)
plt.bar(langs,students)
plt.ylabel('Pengguna')
plt.xlabel('Bahasa Pemrograman')
plt.title('Subplot ke 3')

plt.subplot(2, 2, 4)
plt.barh(y_pos, students)
plt.yticks(y_pos, langs)
plt.xlabel('Pengguna')
plt.ylabel('Bahasa Pemrograman')
plt.title('Subplot ke 4')

plt.tight_layout()
plt.show()
```
