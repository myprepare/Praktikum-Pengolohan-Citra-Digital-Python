## Import Library

```python
import cv2
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
%matplotlib inline
```

## Gray Level Slice



```python
# Load the image
img = data.camera()

# Find width and height of image
row, column = img.shape

# Create an zeros array to store the sliced image
img1 = np.zeros((row,column),dtype = 'uint8')
 
# Specify the min and max range
min_range = 10
max_range = 60
 
# Loop over the input image and if pixel value lies in desired range set it to 255 otherwise set it to 0.
for i in range(row):
    for j in range(column):
        if img[i,j]>min_range and img[i,j]<max_range:
            img1[i,j] = 255
        else:
            img1[i,j] = 0
```

## Plot Image

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title("Citra Input")
ax[1].hist(img.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(img1, cmap=plt.cm.gray)
ax[2].set_title("Citra Output")
ax[3].hist(img1.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```
